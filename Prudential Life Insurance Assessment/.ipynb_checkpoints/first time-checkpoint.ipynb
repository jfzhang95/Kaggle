```python
total = train.isnull().sum().sort_values(ascending=False)
percent = (train.isnull().sum()/train.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

检验剩下的几个特征与response的相关性.

```python
features = ['Family_Hist_4', 'Employment_Info_6', 'Medical_History_1', 'Employment_Info_4', 'Employment_Info_1']
data_check = pd.concat([train[features], train['Response']], axis=1)
corrmat = data_check.corr()
f, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corrmat, vmax=0.8, square=True)
print corrmat
```

可以看到这几个特征与Response的相关性可以忽略,所以可以全部剔除.

```python
drop_features = ['Family_Hist_4', 'Employment_Info_6', 'Medical_History_1', 'Employment_Info_4', 'Employment_Info_1',
                 'Medical_History_10', 'Medical_History_32', 'Medical_History_24', 'Medical_History_15', 'Family_Hist_5',
                'Family_Hist_3', 'Family_Hist_2']
train = train.drop(drop_features, axis=1)
test = test.drop(drop_features, axis=1)
```

```python
train = pd.get_dummies(train)
test = pd.get_dummies(test)
train.head()
```

```python
test.head()
```

```python
corrmat = train.drop('Id',axis=1).corr()
f, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corrmat, vmax=0.8, square=True)
```

```python
k = 15 #number of variables for heatmap
cols = np.abs(corrmat).nlargest(k, 'Response')['Response'].index
cm = np.corrcoef(train[cols].values.T)
sns.set(font_scale=1)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
cols.drop('Response')
```

```python
cols = ['BMI', 'Wt', 'Medical_History_23', 'Medical_Keyword_15', 'Medical_Keyword_3', 'Medical_History_4',
        'Medical_History_39', 'Ins_Age', 'Product_Info_4', 'Medical_Keyword_48', 'Medical_History_6', 'Product_Info_2_D4',
        'Medical_History_16', 'Product_Info_2_D1']
X_train = train[cols]
X_test = test[cols]
y_train = train['Response']
```

```python
from xgboost import XGBClassifier
from sklearn.cross_validation import cross_val_score
```

```python
xgbc = XGBClassifier(learning_rate=5e-2, max_depth=7)
cross_val_score(xgbc, X_train, y_train, cv=5).mean()
```

```python
xgbc.fit(X_train, y_train)
xgbc_y_pred = xgbc.predict(X_test)
```

```python
preds_out = pd.DataFrame({"Id": test['Id'].values, "Response": xgbc_y_pred})
preds_out = preds_out.set_index('Id')
preds_out.to_csv('xgb_offset_submission.csv')
```
